```python
import matplotlib    #hi
matplotlib.use("Agg")
import numpy as np
import matplotlib.pyplot as pt
```

Take the data in "d.txt",

```python
import numpy as np
temp = [];s = ""
ga = "hias"
with open(mode = "r+",file = "d") as d:
    for line in d:
        s+=line
    m = s.split(" ")
    for i in range(0,10000):
        temp.append(m[i][:len(m[i])].split(","))
```

This appropriately changes the text into an array to prepare for analysis, but before that we will make methods to simplify the proccessing of the data

```python
def analyze(arr,age,gen,incBoth):
    t = arr[age][gen].split("/")
    t[0]=int(t[0])+incBoth
    t[1]=int(t[1])+ 1
    arr[age][gen]=str(t[0])+"/"+str(t[1])
def fl(st):
    t = st.split("/")
    return float(t[0])/float(t[1])
```

Now that we have created a method to simplify this process, we can now go on to process the data

```python
percent = []
for i in range(0,99):
    percent.append(["0/0","0/0"])
for num in range(0,len(temp)):
    analyze(percent,int(int(temp[num][0])-1),int(temp[num][1]=="f"),int(temp[num][2]=="t"))       
```

Let's change these fraction to decimal form!

```python
sup = percent
for i in range(0,len(percent)):
    for j in range (0,2):
        sup[i][j] = fl(percent[i][j])
```
